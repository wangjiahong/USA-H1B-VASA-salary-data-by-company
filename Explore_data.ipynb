```python
import seaborn as sns; sns.set_theme(color_codes=True)

sns.set_theme(style="white")
import requests
import pandas as pd
pd.set_option('display.max_rows', None)
import lxml
```

```python
def get_dataframe(company='', jobtitle='Data Scientist', city='', year="All+Years"):
    print(f" ===> Company: {company}")
    print(f" ===> Job Title: {jobtitle}")
    print(f" ===> Year: {year}  \n")
    
    if company=='amazon' and jobtitle =='software engineer':
        jobtitle = 'software development engineer'
    
    url = f"https://h1bdata.info/index.php?em={companies.get(company, company.replace(' ', '+'))}&job={jobtitle.replace(' ', '+')}&city=&year={year}"
    print(url)
    r = requests.get(url)



    html_tables = pd.read_html(r.text, flavor='html5lib')
    df = html_tables[0]
    df = df[~df.EMPLOYER.isnull()].reset_index(drop=True)


    df = df[~df.EMPLOYER.str.contains("adsbygoogle = window.adsbygoogle")].reset_index(drop=True)

    df['BASE SALARY'] = df['BASE SALARY'].astype(int)
    df = df[df['BASE SALARY']<400000]



    print(f"There are {len(df.index)} data points")
    print(f"The median salary is {df['BASE SALARY'].median()}")
    
    
    ax = sns.stripplot(x=df["BASE SALARY"]);
    ax = sns.violinplot(x=df["BASE SALARY"],palette="Set2");
    return df
```

```python
def compare_job_salary_in_same_company(company, jobs = []):
    for jobtitle in jobs:
        print('*'*20)
        print('\n')
        df = get_dataframe(company=company, jobtitle=jobtitle)
```
